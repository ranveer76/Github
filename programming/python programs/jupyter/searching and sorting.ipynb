Code Binary Search

```python
from sys import stdin


def binarySearch(arr, n, x) :
    #Your code goes here
    if x in arr:
        return arr.index(x)
    else:
        return -1


#Taking Input Using Fast I/O
def takeInput() :
    n = int(stdin.readline().strip())
    
    if n == 0 :
        return list(), 0

    arr = list(map(int, stdin.readline().strip().split(" ")))
    return arr, n


#main
arr, n = takeInput()
t = int(stdin.readline().strip())

while t > 0 :
    
    x = int(input().strip())    
    print(binarySearch(arr, n, x))

    t -= 1
```

Code Selection Sort

```python
from sys import stdin

def selectionSort(arr, n) :
    #Your code goes here
    for i in range(1,n):
        j=i-1
        temp=arr[i]
        while(j>=0 and arr[j]>temp):
            arr[j+1]=arr[j]
            j-=1
        arr[j+1]=temp
    return arr, n






#Taking Input Using Fast I/O
def takeInput() :
    n = int(stdin.readline().strip())
    if n == 0 :
        return list(), 0

    arr = list(map(int, stdin.readline().strip().split(" ")))
    return arr, n


#to print the array/list
def printList(arr, n) : 
    for i in range(n) :
        print(arr[i], end = " ")
    print()


#main
t = int(stdin.readline().strip())

while t > 0 :
    
    arr, n = takeInput()
    selectionSort(arr, n)
    printList(arr, n)

    t-= 1
```

Code Merge Two Sorted Arrays

```python
from sys import stdin

def merge(arr1, n, arr2, m) :
    l=[]
    i,j=0,0
    while(i<n and j<m):
        if arr1[i]<arr2[j]:
            l.append(arr1[i])
            i+=1
        else:
            l.append(arr2[j])
            j+=1
    while i<n:
        l.append(arr1[i])
        i+=1
    while j<m:
        l.append(arr2[j])
        j+=1
    return l

#Taking Input Using Fast I/O
def takeInput() :
    n = int(stdin.readline().rstrip())
    if n != 0:
        arr = list(map(int, stdin.readline().rstrip().split(" ")))
        return arr, n

    return list(), 0


#to print the array/list
def printList(arr, n) : 
    for i in range(n) :
        print(arr[i], end = " ")
        
    print()


#main
t = int(stdin.readline().rstrip())

while t > 0 :

    arr1, n = takeInput()
    arr2, m = takeInput()

    ans = merge(arr1, n, arr2, m)
    printList(ans, (n + m))

    t -= 1
```

Push Zeros to end

```python
from sys import stdin

def pushZerosAtEnd(arr, n) :
    #Your code goes here

    s=0
    a=0
    m=[]
    for i in range(n):
        if arr[i]==0:
            s+=1
        else:
            a+=1
            m.append(arr[i])
    for i in range(s):
        m.append(0)
    return m

#Taking Input Using Fast I/O
def takeInput() :
    n = int(stdin.readline().rstrip())

    if n == 0:
        return list(), 0
    
    arr = list(map(int, stdin.readline().rstrip().split()))
    return arr, n
  

#to print the array/list
def printList(m, n) : 
    for i in range(n) :
        print(m[i], end = " ")

    print()


#main
t = int(stdin.readline().strip())

while t > 0 :

    arr, n = takeInput()

    m=pushZerosAtEnd(arr, n)
    printList(m, n)

    t -= 1
```

Rotate array

```python
from sys import stdin


def rotate(arr, n, d):
    #Your code goes here
    m=[]
    j=[]
    for i in range(n):
        if i<d:
            j.append(arr[i])
        else:
            m.append(arr[i])
    m+=j
    return m

#Taking Input Using Fats I/O
def takeInput() :
    n = int(stdin.readline().rstrip())
    if n == 0:
        return list(), 0

    arr = list(map(int, stdin.readline().rstrip().split(" ")))
    return arr, n


#to print the array/list 
def printList(arr, n) : 
    for i in range(n) :
        print(arr[i], end = " ")
    print()


#main
t = int(stdin.readline().rstrip())

while t > 0 :
    
    arr, n = takeInput()
    d = int(stdin.readline().rstrip())
    arr=rotate(arr, n, d)
    printList(arr, n)
    
    t -= 1
```

Second Largest in array

```python
from sys import stdin
def secondLargestElement(arr, n):
    if n==0 or arr[0]==arr[1] and arr[1]==arr[2] and arr[2]==arr[3]:
        return -2147483648
    k=max(arr)
    arr[arr.index(k)]=-9
    l=max(arr)
    while l==k:
        arr[arr.index(l)]=-9
        l=max(arr)
    return l
def takeInput() :
    n = int(stdin.readline().rstrip())
    if n != 0:
        arr = list(map(int, stdin.readline().rstrip().split(" ")))
        return arr, n
    return list(), 0
t = int(stdin.readline().rstrip())
while t > 0 : 
    arr, n = takeInput()
    print(secondLargestElement(arr, n))
    t -= 1
```

Check Array Rotation

```python
from sys import stdin

def arrayRotateCheck(arr, n):
    #Your code goes here
    s=1
    for i in range(n-1):
        if arr[i]<arr[i+1]:
            s+=1
        else:
            break
    if s<n and s>1:
        return s
    else:
        return 0


#Taking Input Using Fast I/O
def takeInput() :
    n = int(stdin.readline().rstrip())
    if n == 0:
        return list(), 0

    arr = list(map(int, stdin.readline().rstrip().split(" ")))
    return arr, n


#main
t = int(stdin.readline().rstrip())

while t > 0 :

    arr, n = takeInput()
    print(arrayRotateCheck(arr, n))

    t -= 1
```
