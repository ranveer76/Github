Row Wise Sum

```python
from sys import stdin

def rowWiseSum(mat, nRows, mCols):
    #Your code goes here
    for i in range(nRows):
        s=0
        for j in range(mCols):
            s+=mat[i][j]
        print(s,end=' ')


#Taking Input Using Fast I/O
def take2DInput() :
    li = stdin.readline().rstrip().split(" ")
    nRows = int(li[0])
    mCols = int(li[1])
    
    if nRows == 0 :
        return list(), 0, 0
    
    mat = [list(map(int, input().strip().split(" "))) for row in range(nRows)]
    return mat, nRows, mCols


#main
t = int(stdin.readline().rstrip())

while t > 0 :

    mat, nRows, mCols = take2DInput()
    rowWiseSum(mat, nRows, mCols)
    print()

    t -= 1
```

Largest Row or Column

```python
from sys import stdin

def findLargest(arr, nRows, mCols):
    #Your code goes here
    if nRows==0 and mCols==0:
        print('row',0,-2147483648)
    elif nRows==mCols:
        ro=[]
        co=[]
        for i in range(nRows):
            r=0
            c=0
            for j in range(mCols):
                c+=arr[j][i]
                r+=arr[i][j]
            co.append(c)
            ro.append(r)
        k=max(ro)
        L=max(co)
        if k==L or k>L:
            print('row',ro.index(k),k)
        elif k<L:
            print('column',co.index(L),L)
    else:
        ro=[]
        co=[]
        for i in range(nRows):
            r=0
            for j in range(mCols):
                r+=arr[i][j]
            ro.append(r)
        for i in range(mCols):
            c=0
            for j in range(nRows):
                c+=arr[j][i]
            co.append(c)
        L=max(co)
        k=max(ro)
        if k>=L:
            print('row', ro.index(k), k)
        else:
            print('column',co.index(L),L)

#Taking Input Using Fast I/O
def take2DInput() :
    li = stdin.readline().rstrip().split(" ")
    nRows = int(li[0])
    mCols = int(li[1])
    
    if nRows == 0 :
        return list(), 0, 0
    
    mat = [list(map(int, input().strip().split(" "))) for row in range(nRows)]
    return mat, nRows, mCols


#main
t = int(stdin.readline().rstrip())

while t > 0 :

    mat, nRows, mCols = take2DInput()
    findLargest(mat, nRows, mCols)

    t -= 1
```

Wave Print

```python
from sys import stdin

def wavePrint(mat, nRows, mCols):
    #Your code goes here
    i=0
    t=0
    b=nRows
    while i<mCols:
        if i%2==0:
            x=t
            y=nRows
            z=1
        else:
            x=b-1
            y=-1
            z=-1
        for j in range(x,y,z):
            print(mat[j][i],end=' ')
        i+=1

#Taking Iput Using Fast I/O
def take2DInput() :
    li = stdin.readline().rstrip().split(" ")
    nRows = int(li[0])
    mCols = int(li[1])
    
    if nRows == 0 :
        return list(), 0, 0
    
    mat = [list(map(int, input().strip().split(" "))) for row in range(nRows)]
    return mat, nRows, mCols


#main
t = int(stdin.readline().rstrip())

while t > 0 :

    mat, nRows, mCols = take2DInput()
    wavePrint(mat, nRows, mCols)
    print()

    t -= 1
```

Spiral Print

```python
from sys import stdin

def spiralPrint(mat, nRows, mCols):
    #Your code goes here
    top=0
    bottom=nRows
    left=0
    right=mCols
    i=0
    z=0
    z1=1
    z2=2
    z3=3
    while i<=(nRows*2)-1:
        if i==z:
            for j in range(left,right):
                print(mat[top][j],end=' ')
            top+=1
            z+=4
        elif i==z1:
            for j in range(top,bottom):
                print(mat[j][right-1],end=' ')
            right-=1
            z1+=4
        elif i==z2:
            for j in range(right-1,left-1,-1):
                print(mat[bottom-1][j],end=' ')
            bottom-=1
            z2+=4
        elif i==z3:
            for j in range(bottom-1,top-1,-1):
                print(mat[j][left],end=' ')
            left+=1
            z3+=4
        i+=1

#Taking Input Using Fast I/O
def take2DInput() :
    li = stdin.readline().rstrip().split(" ")
    nRows = int(li[0])
    mCols = int(li[1])
    
    if nRows == 0 :
        return list(), 0, 0
    
    mat = [list(map(int, input().strip().split(" "))) for row in range(nRows)]
    return mat, nRows, mCols


#main
t = int(stdin.readline().rstrip())

while t > 0 :

    mat, nRows, mCols = take2DInput()
    spiralPrint(mat, nRows, mCols)
    print()

    t -= 1
```
